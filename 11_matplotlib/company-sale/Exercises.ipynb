# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/refs/heads/main/data/matplotlib/company_sales_data.csv"
df = pd.read_csv(url, sep ="\t")
print(df)
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python

plt.figure(figsize=(8, 5))
total_profit = df.groupby('month_number')['total_profit'].sum()
plt.plot(total_profit.index, total_profit.values)
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.ylim(100000, 500000)
plt.xticks(total_profit.index)

plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
plt.figure(figsize=(8, 5))
x = df['month_number']
y = df['total_profit']
plt.plot(x, y, color='r', linestyle='dashed',marker='o', linewidth =5, markerfacecolor="black",markersize ="8", label="Profit data of last year ")
plt.xticks(x.index)
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.ylim(100000, 500000)
plt.legend(loc ="lower right")
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
plt.figure(figsize=(10, 6))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    plt.plot(df['month_number'], df[product], marker='o', label=product+"sales Data",linewidth="5",markersize ="8")
plt.title("Sales data of all products")
plt.xlabel("Month number")
plt.ylabel("Sales units")
plt.xticks(df['month_number'])
plt.legend(loc='upper left')
plt.ylim(0,18000)
plt.show()

```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8, 5))
plt.scatter(df['month_number'], df['toothpaste'], color='#4682B4', marker='o', label='Toothpaste')
plt.title("Toothpaste sales data")
plt.xlabel("Month number")
plt.ylabel("Sales units")
plt.xticks(df['month_number'])
plt.legend()
plt.grid(visible=True, linestyle = "--")

plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
plt.figure(figsize=(10, 6))
x = df['month_number']
width = 0.25
plt.bar(x - width/2, df['facecream'], width=width, label='Face cream sales data')
plt.bar(x + width/2, df['facewash'],  width=width, label='Face wash sales data')
plt.title("Facecream and Facewash sales data")
plt.xlabel("Month number")
plt.ylabel("Sales units")
plt.xticks(x)
plt.legend(loc ="upper left")
plt.grid(visible=True, linestyle ="--")
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/06.png?raw=1)

```python
plt.figure(figsize=(8, 5))
plt.bar(df['month_number'], df['bathingsoap'], color='#4682B4')
plt.title("Bathing soap sales data")
plt.xlabel("Month number")
plt.ylabel("Sales units")
plt.xticks(df['month_number'])
plt.legend()
plt.grid(visible=True, linestyle = "--")
plt.savefig("bathing_soap_sales.png", dpi=150)
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
plt.figure(figsize=(8, 5))
plt.hist(df['total_profit'], bins=6, color='purple', edgecolor='black')
plt.title("Profit range histogram")
plt.xlabel("Profit in dollar")
plt.ylabel("Frequency")
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
plt.figure(figsize=(8, 8))
totals = df[products].sum()
plt.pie(totals, labels=products, autopct='%1.1f%%')
plt.title("Total sales by product (last year)")
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 5))

ax1.plot(df['month_number'], df['bathingsoap'], color='black', marker='o',linewidth ="5",markersize="8")
ax1.set_title("Bathing soap sales")
ax1.set_xlabel("Month number")
ax1.set_ylabel("Sales units")
ax1.set_xticks(df['month_number'])

ax2.plot(df['month_number'], df['facewash'], color='r', marker='o',linewidth="5",markersize ="8")
ax2.set_title("Facewash sales")
ax2.set_xlabel("Month number")
ax2.set_ylabel("Sales units")
ax2.set_xticks(df['month_number'])

plt.suptitle("Bathing soap and Facewash sales data")
plt.tight_layout()
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
plt.figure(figsize=(10, 6))
plt.stackplot(df['month_number'],
              [df[p] for p in products],
              labels=products)
plt.title("All product sales - Stack plot")
plt.xlabel("Month number")
plt.ylabel("Sales units")
plt.xticks(df['month_number'])
plt.legend(loc='upper left')
plt.show()
```
